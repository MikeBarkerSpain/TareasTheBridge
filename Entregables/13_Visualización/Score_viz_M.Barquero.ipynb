### Step 1. Import the necessary libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import scipy.stats as stats
import seaborn as sns
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
# creaci√≥n del df en base a un diccionario
dict_ = {'first_name':['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 'last_name':['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 'age':[42,52,36,24,73], 'female':[0,1,1,0,1], 'preTestScore':[4,24,31,2,3],'postTestScore':[25,94,57,62,70]}

df= pd.DataFrame(dict_)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
# se genera el plot vac√≠o y se completan los argumentos requeridos
fig, ax = plt.subplots()
ax.set_title('preTestScore vs postTestScore')
# se genera el visualizador necesario
sns.scatterplot(data=df, x='preTestScore', y='postTestScore', size='age', sizes=(20,200))
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
fig, ax = plt.subplots()
ax.set_title('preTestScore VS postTestScore')
ax.legend().remove()
df['postTestScore4.5']=df.postTestScore.apply(lambda x: x*4.5)
sns.scatterplot(data=df, x='preTestScore', y='postTestScore', size='postTestScore4.5', hue='female', sizes=(50,300))
```
