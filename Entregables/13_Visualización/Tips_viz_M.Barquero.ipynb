### Step 1. Import the necessary libraries:

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import scipy.stats as stats
import seaborn as sns
from mpl_toolkits.axes_grid1 import make_axes_locatable
# print the graphs in the notebook
# % matplotlib inline
sns.set_context('notebook')


# set seaborn style to white
sns.set_style("white")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv', sep=',')
df
```

### Step 3. Assign it to a variable called tips

```python
tips = df
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop(columns='Unnamed: 0', inplace=True)
tips
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(tips, x='total_bill', stat='frequency', kde=True)
```

### Step 7.  Create one image with the relationship of `total_bill`, `tip` and `size`.
#### Hint: It is just one function.

```python
x_vars = ["total_bill", "tip", "size"]
y_vars = x_vars
GridPlot = sns.PairGrid(tips, x_vars=x_vars, y_vars=y_vars)
GridPlot.map_diag(sns.histplot)
GridPlot.map_offdiag(sns.scatterplot)
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.scatterplot(data=tips, x="day", y="total_bill", palette="set2")
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(data=tips, x="tip", y="day", palette="deep", hue='sex')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
# Cómo aumentar el tamaño de un plot 
fig_dims = (8, 6)
fig, ax = plt.subplots(figsize=fig_dims)

pltbox = sns.boxplot(x="day", y="total_bill", data=tips, hue='time', order=['Thur', 'Fri', 'Sat', 'Sun'], ax=ax)
leg = pltbox.legend( loc = 'upper left')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
#tipval = sns.catplot(x="tip", col="time", col_wrap=2, data=tips, kind="count", height=5, aspect=.8)
#sns.histplot(data=tips, x='tip', col="time", col_wrap=2)

dinner_tip = tips[tips['time'] == 'Dinner']['tip'].dropna().to_list()
lunch_tip = tips[tips['time'] == 'Lunch']['tip'].dropna().to_list()

fig, axs = plt.subplots(1, 2, tight_layout=False, sharey='row')
axs[0].hist(dinner_tip, bins=12)
axs[1].hist(lunch_tip, bins=12)
```

```python
tip_hist = sns.FacetGrid(tips, col="time")
tip_hist.map(sns.histplot, "tip")
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
scat_smok_gen = sns.FacetGrid(tips, col="sex", hue="smoker")
scat_smok_gen.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
scat_smok_gen.add_legend()
```
