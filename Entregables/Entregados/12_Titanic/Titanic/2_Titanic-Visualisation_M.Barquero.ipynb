13. Using Seaborn plot the correlation matrix of various features.

Hint: search how to use `heatmap`



```python
#your code here: modification of the original titanic table to get interesting data
titanic_data = titanic.drop(columns=['name', 'sibsp','parch','ticket','embarked','boat','home.dest'])
titanic_data['passenger'] = 1
```

```python
# correlation between class an survival
titanic_data_heatmap = titanic_data.drop(columns=['age', 'gender', 'fare'])
titanic_data_heatmap = titanic_data_heatmap.groupby(['pclass', 'survived']).sum()
titanic_data_heatmap = titanic_data_heatmap.reset_index().pivot('pclass', 'survived', 'passenger')
sns.heatmap(titanic_data_heatmap, linewidths=.5)
```

```python
# correlation between fare and survival
titanic_data_heatmap = titanic_data.drop(columns=['age', 'gender', 'pclass'])
titanic_data_heatmap['fare'] = titanic_data_heatmap['fare'].apply(lambda x: x//20*20)
titanic_data_heatmap = titanic_data_heatmap.groupby(['fare', 'survived']).sum()
titanic_data_heatmap = titanic_data_heatmap.reset_index().pivot('fare', 'survived', 'passenger')
sns.heatmap(titanic_data_heatmap, linewidths=.5)
```

```python
titanic_data_heatmap = titanic_data.drop(columns=['survived', 'gender', 'pclass'])
titanic_data_heatmap['fare'] = titanic_data_heatmap['fare'].apply(lambda x: x//20*20)
titanic_data_heatmap['age'] = titanic_data_heatmap['age'].apply(lambda x: x//5*5)
titanic_data_heatmap = titanic_data_heatmap.groupby(['fare', 'age']).sum()
titanic_data_heatmap = titanic_data_heatmap.reset_index().pivot('fare', 'age', 'passenger')
sns.heatmap(titanic_data_heatmap, linewidths=.5)
```
