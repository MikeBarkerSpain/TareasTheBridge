### ReLu

```python
def ReLU(x):
    return (abs(x) + x) / 2
```

```python
ReLU(32626)
```

### Sigmoid

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(-590)
```
