### los pasos a seguir para resolver un ejercicio de regresión lineal simple

1.- Cargamos el conjunto de datos
2.- Partimos el conjunto de train y test
3.- Creamos una variable que represente el modelo (inicializamos)
4.- Fit (Entrenamos) el modelo con el conjunto train
5.- Calculamos el error de train y test con respecto a los valores predichos
6.- ¿Es un buen error? -> si no, repetir el proceso
    6.1.- Cambiando la semilla
    6.2.- Eligiendo un porcentaje de test diferente
    6.3.- Coger más datos/otra muestra
