### Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión lineal entre cada columna numérica con la columna 'price'. 

Cuando existe una columna(variable) que es la que se quiere comparar, se le llama 'target'. 

```python
import pandas as pd

USA_Housing = pd.read_csv("../data/USA_Housing.csv")
USA_Housing
```

```python
import numpy as np 
from sklearn.model_selection import train_test_split
#importing the Linear Regression Algorithm
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn import metrics
```

```python
# Precio es siempre la variable dependiente
y = USA_Housing['Price']
y = np.array(y).reshape(-1, 1)

# se selecciona la variable independiente y se manipula para utilizar la función
X = USA_Housing['Avg. Area Income']
X = np.array(X).reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

```

```python
# Precio es siempre la variable dependiente
y = USA_Housing['Price']
y = np.array(y).reshape(-1, 1)

# se recorre el dataframe para examinar cada variable independiente y se manipula para utilizar la función
for i in range(len(USA_Housing.columns)):
    if USA_Housing.dtypes[i] == 'float64' and USA_Housing.columns[i] != 'Price':
        X = USA_Housing[USA_Housing.columns[i]]
        X = np.array(X).reshape(-1, 1)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

        #Creación del objeto LinearRegression
        lm = LinearRegression(n_jobs=-1)  # -1 para usar todos los procesadores
        lm.fit(X_train, y_train)    #Entrenar el modelo

        # Output resultados
        print("\n\n\nResultados de la regresión para "+ USA_Housing.columns[i])

        # Obtener las métricas de control 
        print("Origen de la regresión:", lm.intercept_)
        print("Pendiente de la regresión:", lm.coef_)
        lm.score(X_train, y_train) *100
        lm.score(X_test, y_test) *100

        # Predecir el conjunto de valor en función de la muestra de test
        predictions = lm.predict(X_test)

        # Control de calidad de la predicción 
        print('MAE:', metrics.mean_absolute_error(y_test, predictions))
        print('MSE:', metrics.mean_squared_error(y_test, predictions))
        print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
        print(f"Porcentaje de acierto en entrenamiento:"+str(round(lm.score(X_train, y_train), 2))+"%.")
        print(f"Porcentaje de acierto en prueba:"+str(round(lm.score(X_test, y_test), 2))+"%.")
        print(f"Porcentaje de acierto total:"+str(round(lm.score(X, y), 2))+"%.")

        plt.scatter(y_test, predictions, edgecolor='black')
        plt.show()
```

# Código de test


```python
#Creación del objeto LinearRegression
lm = LinearRegression(n_jobs=-1)  # -1 para usar todos los procesadores
lm.fit(X_train, y_train)    #Entrenar el modelo

# Obtener las métricas de control 
print("Origen de la regresión:", lm.intercept_)
print("Pendiente de la regresión:", lm.coef_)
lm.score(X_train, y_train) *100
lm.score(X_test, y_test) *100

# Predecir el conjunto de valor en función de la muestra de test
predictions = lm.predict(X_test)

# Control de calidad de la predicción 
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print(f"Porcentaje de acierto en entrenamiento:"+str(round(lm.score(X_train, y_train), 2))+"%.")
print(f"Porcentaje de acierto en prueba:"+str(round(lm.score(X_test, y_test), 2))+"%.")
print(f"Porcentaje de acierto total:"+str(round(lm.score(X, y), 2))+"%.")
```
