A partir del archivo food_100, se pide:

1. Limpiar el archivo borrando las columnas Unnamed.
2. ¿Qué porcentaje de valores NaN hay en cada columna?
3. ¿Tendría algún sentido clasificar el nombre de los alimentos a partir del top5 de columnas numéricas con menos valores NaN?

No tiene sentido ya que los datos de las columnas code, created_datetime, last_modified_t y last_modified_datetime son de origen técnico y no incorporan información del producto

```python
import pandas as pd
df = pd.read_csv("../data/food_100.csv")
rows= df.shape[0]
df = df.drop(columns=['Unnamed: 0'])

for i in range(len(df.keys())):
    por_nan_col = df.isna().sum()[i]/rows
    print("El porcentaje de nans en "+df.keys()[i]+ " es " + str(por_nan_col) + '%')
```
